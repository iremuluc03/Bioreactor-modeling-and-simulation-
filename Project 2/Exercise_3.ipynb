```python
# d)

#imports

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# constants
k1b = 10.0      # s^-1
k2f = 22.0      # s^-1
KmS = 210.0     # uM
KmP = 97.0      # uM
E0 = 0.1        # mol/L 
S0 = 1          # mol/L
P0 = 0          # mol/L


k1f = (k1b + k2f) / KmS  # uM^-1 s^-1
k2b = (k1b + k2f) / KmP  # uM^-1 s^-1

# ODE
def PGM_system(t, y):
    S, P = y
    
    # reaction rate v
    numerator = E0 * (k1f * k2f * S - k1b * k2b * P)
    denominator = (k1b + k2f) + k1f * S + k2b * P
    v = numerator / denominator
    
    dSdt = -v
    dPdt = v
    return [dSdt, dPdt]


y0 = [S0, P0]
t_span = (0, 100)  
t_eval = np.linspace(0, 100, 500)  

# solve
solution = solve_ivp(PGM_system, t_span, y0, t_eval=t_eval, method='RK45')

t_res = solution.t
S_res = solution.y[0]
P_res = solution.y[1]

# plot
plt.figure(figsize=(9, 5))
plt.plot(t_res, S_res, label='[S] (3-phospho-D-glycerate)')
plt.plot(t_res, P_res, label='[P] (2-phospho-D-glycerate)')

plt.title('Reversible Reaction Kinetics (QSSA Model)')
plt.xlabel('Time (s)')
plt.ylabel('Concentration (mol/L)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

keq_kinetic = (k1f * k2f) / (k1b * k2b)
final_ratio = P_res[-1] / S_res[-1]
print(f"Theoretical Keq: {keq_kinetic:.4f}")
print(f"Final simulated [P]/[S] ratio: {final_ratio:.4f}")
```

```python
# e)

P0_e = 1          # mol/L
y0_e = [S0, P0_e]

solution_e = solve_ivp(PGM_system, t_span, y0_e, t_eval=t_eval, method='RK45')

t_res_e = solution_e.t
S_res_e = solution_e.y[0]
P_res_e = solution_e.y[1]

# plot
plt.figure(figsize=(9, 5))
plt.plot(t_res_e, S_res_e, label='[S] (3-phospho-D-glycerate)')
plt.plot(t_res_e, P_res_e, label='[P] (2-phospho-D-glycerate)')

plt.title('Reversible Reaction Kinetics (QSSA Model)')
plt.xlabel('Time (s)')
plt.ylabel('Concentration (mol/L)')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()

final_ratio_e = P_res_e[-1] / S_res_e[-1]
print(f"Theoretical Keq: {keq_kinetic:.4f}")
print(f"Final simulated [P]/[S] ratio: {final_ratio:.4f}")
```
