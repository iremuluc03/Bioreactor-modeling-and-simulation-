```python
## exo f)

import matplotlib.pyplot as plt
import numpy as np  
import math as m
from scipy.integrate import solve_ivp

D = np.linspace(0, 1, 100)
alpha = [0.1, 0.5, 0.99]
C=2
mu_max=1
Y_xs=0.5
F=1500
S0=2
Ks=0.01

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

for a in alpha:
    ## Calcul of D_max
    D_max=mu_max/(1+a*(1-C))
    ## Calcul Cell concentration 
    mu=D*(1+a*(1-C))

    S=Ks*D*(1+a*(1-C))/(mu_max-D*(1+a*(1-C)))
    
    X1=Y_xs*(S0-S)/(1-a*(1-C))
    ## Calcul of Productivity 
    X2=(1+a)*X1-a*C*X1
    P=X2*D

    # On trace X1 sur le premier axe
    ax1.plot(D, X1, label='alpha={}'.format(a))
    # On trace P sur le second axe
    ax2.plot(D, P, label='alpha={}'.format(a))

ax1.set_title("Biomasse X1")
ax1.set_xlabel("D")
ax1.legend()

ax2.set_title("Productivité P")
ax2.set_xlabel("D")
ax2.legend()
    
    
    






```
