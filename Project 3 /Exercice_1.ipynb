```python
## exo f)

import matplotlib.pyplot as plt
import numpy as np  
import math as m
from scipy.integrate import solve_ivp


## Paramètres ##

C=2
mu_max=1
Y_xs=0.5
F=1500
S0=2
Ks=0.01

D = np.linspace(0.001, 11, 500)

alpha_list = [0.1, 0.5, 0.9]


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

for a in alpha_list:
    
    ## Calcul of D_max
    D_max=mu_max/(1+a*(1-C))
    
    ## Calcul S 
    
    S=Ks*D*(1+a*(1-C))/(mu_max-D*(1+a*(1-C)))
    
    ## Calcul of Biomass X1

    X1=Y_xs*(S0-S)/(1+a*(1-C))

    ## Calcul of Productivity 
    X2=(1+a)*X1-a*C*X1
    P=X2*D

    ## Masquage du washout pour les valeurs de D > D_max

    mask= (D<D_max) & (S<S0)
    X1_plot = np.where(mask, X1, 0)
    P_plot = np.where(mask, P, 0)

    # On trace X1 sur le premier axe
    ax1.plot(D, X1_plot, label='alpha={}'.format(a))
    # On trace P sur le second axe
    ax2.plot(D, P_plot, label='alpha={}'.format(a))

ax1.set_title("Biomasse X1")
ax1.set_xlabel("D")
ax1.legend()

ax2.set_title("Productivité P")
ax2.set_xlabel("D")
ax2.legend()
    
    
    






```
